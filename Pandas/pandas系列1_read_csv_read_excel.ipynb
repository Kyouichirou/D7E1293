## 参数解读

由于参数太多, 只解读相对常用的

1. 通用
filepath_or_buffer, 这个不需要解释, 读取文件的路径, 当然不限于本地的路径, 在线文档也可以, 例如存放在github上的文件

usecols, 指定需要读取的列, 需要注意和read_Excel, 在支持格式上是有所差异的

names, 在读取完成时就对列名称进行修改, 注意和usecols同时使用的冲突问题

sep=NoDefault.no_default

delimiter=None, header='infer'
这两个参数用于指定分隔符不是逗号, 而是其他符号使用

2. 读取格式相关
dtype, 列数据以什么格式进行读取, 在数据特别大时, 假如可以指定数据的读取格式, 理论上有助于加快数据的读取, 字符串的读取方式的实用性最广, 需要注意float, int数据格式对于空值的处理问题

3. 警告相关
encoding_errors='strict',

4. 内存管理相关
iterator
chunksize,  即分块读取, 用于处理大型文件
low_memory

5. 符号相关
这两个参数需要注意一个问题, 在处理跨语言的文件时, 详情参阅下面的图片.
thousands, 千分位分隔符
decimal, 小数点分隔符

6. 转换处理相关
keep_default_na, 对待空值的处理方式, 默认将空值转为nan值, 这个参数需要注意, 影响非常广在表格的处理上, 这涉及到nan值的运算的特性 converters=None
true_values=None
false_values=None
这些在对读取数据对数字, 空值等数据进行转换

7. 读取内容限制相关
header, 指定表头所在行的位置, 例如表头的位置在第二行, header=1(header是从0开始算起)

8. 高级
engine, 指定读取数据的引擎, 多数情况下"C"引擎就可以(默认值), 但是在跨语言的处理上, 执行某些数据清洗时, 就需要功能更完善的"Python"引擎

目前支持3中engine, 默认的C, 注意csv的engine和read_Excel是不一样的.

由于在参数上部分是和read_csv高度重叠的, 需要注意的是csv文件和Excel文件的差异引发的一些问题

csv文件是没有多个sheet的概念的, 所以excel涉及的sheet, 在csv不适用

Excel的engine也需要注意, 和是csv的差异

需要注意在使用read_excel时, 可能会出现没有指定格式的warning警告, 这种警告一般出现在Java程序生成的Excel文件上.
